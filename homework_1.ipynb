```python
#23  Write a function that takes in any word and counts the number of a particular letter in it. Give the function a default value if no word is passed in
def count_letter(word="Andrew", letter="e"):
    count = 0
    for char in word:
        if char == letter:
            count += 1
    return count
count_letter("elephant"), count_letter("Moritz")


```
