```python
import pandas as pd

```

```python
belize_listings_path = "belize_airbnb_listings.csv"
belize_df = pd.read_csv(belize_listings_path)
belize_df.columns
```

```python
belize_neighborhoods_df= belize_df.groupby('neighbourhood')
belize_neighborhoods_df=belize_neighborhoods_df.agg({
    'price':'mean',
    'number_of_reviews':'sum',
    'availability_365': 'max'
})
print(belize_neighborhoods_df)
```
