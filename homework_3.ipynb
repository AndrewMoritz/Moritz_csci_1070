```python
#Starter code. Do NOT modify this cell
import numpy as np
import pandas as pd

# Read in dataset
df = pd.read_csv('cars.csv',delimiter=';')

# Convert each column to an np.array
car = df.Car.to_numpy()
mpg = df.MPG.to_numpy()
cylinders = df.Cylinders.to_numpy()
displacement = df.Displacement.to_numpy()
horsepower = df.Horsepower.to_numpy()
weight = df.Weight.to_numpy()
acceleration = df.Acceleration.to_numpy()
model = df.Model.to_numpy()
origin = df.Origin.to_numpy()

# create an array of headers
headers = np.array(['Car', 'MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight','Acceleration', 'Model', 'Origin'])
```

```python
# 10. What is the correlation coefficient of mpg and weight?

print(np.corrcoef(mpg, weight))
# The correlation coefficient is -.7875

# What is a correlation coefficient? 
# A correlation coefficient is a measurement of how strong a relationship between two variables with a coefficient of 0 meaning no relationship, and 1 or -1 meaning a perfect relationship
# What does the result indicate about the relationship between mpg and weight?
# This result indicates that there is a decently strong negative relationship betweeen weight and mpg, meaning that as weight increases mpg decreases
```
